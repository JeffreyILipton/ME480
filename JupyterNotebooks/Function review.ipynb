```python
def summationCube(n):
    total, k = 0, 1
    while k <= n:
        total, k = total + pow(k,3), k + 1
    return total

def summationOneOver(n):
    total, k = 0, 1
    while k <= n:
        if k>0:
            total, k = total + 1/k, k + 1
        else:
            total, k = total + 0, k + 1
    return total
```

```python
summationCube(10)
```

```python
summationOneOver(10)
```

```python
def summation(n, term):
    total, k = 0, 1
    while k <= n:
        total, k = total + term(k), k + 1
    return total

def cube(k):
    return pow(k,3)

def oneOver(k):
    if k==0: return 0
    return 1/k

```

```python
summation(10,cube)
```

```python
summation(10,oneOver)
```

# Return functions

```python
def incrementListBy(n):
    def adder(listin):
        y=[]
        for a in listin:
            y.append(a+n)
        return y
    return adder
```

```python
iL1 = incrementListBy(3)
```
