Unlike lists or tuples, the contents of a numpy array **MUST** be the same type.
\

the type is generally an integer, float or string. But it can be any object as well
\ 

You will see int32 or float64. the number refers to the number of bytes the data is stored as.

```python
import timeit
a = np.array(['abba' for _ in range(10000)])
b = np.array(['abba' for _ in range(10000)], dtype=object)
```

```python
%timeit a.copy()
%timeit b.copy()
```

```python
import timeit
a = np.array([i for i in range(10000)])
b = np.array([i for i in range(10000)], dtype=object)
%timeit a.copy()
%timeit b.copy()
```

## Combining Arrays

```python
a = np.array([[9., 7.], [5., 2.]])
b = np.array([[1., 9.], [5., 1.]])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
A = np.eye(2) * 2
B = np.eye(3) * 3
np.block([

    [A,               np.zeros((2, 3))],

    [np.ones((3, 2)), B               ]

])
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6]])
```

```python
np.concatenate((a, b), axis=0)
```

```python
np.concatenate((a, b.T), axis=1)
```

```python
np.concatenate((a, b), axis=None)
```

Slices and reshapings are actually just **views** of the same underlying data structure

```python
x = np.array([[1,2],[3,4]])
c = x.view()
print(c)
print(x)
```

```python
c[0,0]=0
print(x)
```

```python
c is x
```

```python
c.base is x
```

```python
x = np.array([[1,2],[3,4]])
c = x.copy()
c[::]=0
print(x)
print(c)
```

## Matrix Operations

```python
A = np.array(  [[1,1],
                [0,1]] )
```

```python
B = np.array( [[2,0],
               [3,4]] )
```

```python
A@B
```

```python
A.dot(B)
```

```python
np.linalg.det(A)
```

```python
np.linalg.inv(A)
```

Solve $A \vec{X}=\vec{B}$

```python
A = np.array([[3,1], [1,2]])
B = np.array([9,8])
X = np.linalg.solve(A, B)
print(X)
```

```python
A = np.diag((1, 2, 3))
```

```python
np.linalg.eigvals(A)
```

```python
np.linalg.eig(A)
```
